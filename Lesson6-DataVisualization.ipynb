## Simple Graph
Before anything else, always import matplotlib.pyplot. After importing, the function PLOT is now available for use

```python
import matplotlib.pyplot as plt

x_numbers = [1,2,3]
y_numbers = [2,4,6]

plt.plot(x_numbers, y_numbers)

plt.show()
```

## Marking Points on Your Graph
If you want the graph to mark the points that you supplied for plotting, you can use an additional keyword argument while calling the plot() function

```python
import matplotlib.pyplot as plt

x_numbers = [1,2,3]
y_numbers = [2,4,6]

plt.plot(x_numbers, y_numbers, marker = "o")

plt.show()
```

## Comparing the trends
You can also plot more than 1 line in a single graph to make a comparison of trends.

```python
import matplotlib.pyplot as plt

temp_2000 = [31.3,37.3,47.2,51.0,63.5,71.3,72.3,72.7,66.0,57.0,45.3,31.1]
temp_2001 = [40.9,35.7,43.1,55.7,63.1,71.0,77.9,75.8,66.6,56.2,51.9,43.6]
temp_2002 = [37.3,40.9,50.9,54.8,65.1,71.0,78.8,76.7,68.8,58.0,43.9,41.5]

months = range(1,13)
plt.plot(months,temp_2000,months,temp_2001,months,temp_2002)

plt.show()
```

## Adding a legend box
When doing data visualization, a legend box is necessary to inform the users of the representation of the data in a particular visuals. When adding a legend box, use plt.legend(["label1","label2","label3"])

```python
import matplotlib.pyplot as plt

temp_2000 = [31.3,37.3,47.2,51.0,63.5,71.3,72.3,72.7,66.0,57.0,45.3,31.1]
temp_2001 = [40.9,35.7,43.1,55.7,63.1,71.0,77.9,75.8,66.6,56.2,51.9,43.6]
temp_2002 = [37.3,40.9,50.9,54.8,65.1,71.0,78.8,76.7,68.8,58.0,43.9,41.5]

months = range(1,13)
plt.plot(months,temp_2000,months,temp_2001,months,temp_2002)
plt.legend([2000,2001,2002])

plt.show()
```

## Adding a Title and Labels
When adding a title to the graph use the plt.title() function and add labels for the x- and y-axes using the plt.xlabel() and plt.ylabel() functions.

```python
import matplotlib.pyplot as plt

temp_2000 = [31.3,37.3,47.2,51.0,63.5,71.3,72.3,72.7,66.0,57.0,45.3,31.1]
temp_2001 = [40.9,35.7,43.1,55.7,63.1,71.0,77.9,75.8,66.6,56.2,51.9,43.6]
temp_2002 = [37.3,40.9,50.9,54.8,65.1,71.0,78.8,76.7,68.8,58.0,43.9,41.5]

months = range(1,13)
plt.plot(months,temp_2000,months,temp_2001,months,temp_2002)
plt.title("Monthly Philippine Temperature 2000 - 2002")
plt.ylabel("Temperature")
plt.xlabel("Months")
plt.legend(["2000","2001","2002"])


plt.show()
```

## More informative line chart example

```python
import matplotlib.pyplot as plt

years = [1950,1960,1970,1980,1990,2000,2010]
gdp = [300.2, 543.3, 1075.9, 2862.5, 5979.6, 10289.7, 14958.3]

#create a line chart, years on x-axis, gdp on y-axis
plt.plot(years, gdp, color = 'green', marker = 'o', linestyle = 'solid')

#add a title
plt.title("Nominal GDP")

#add a label to the y-axis
plt.ylabel("Billions of $")

plt.show()
```

## Bar Charts

A bar chart is a good choice when you want to show how some quantity varies among some discrte set of items. For example, let us show the number of men's basketball championships at the UAAP.

```python
schools = ["ADU", "ADMU", "DLSU", "FEU", "NU", "UE", "UST", "UP"]
num_champs = [1,10,9,20,2,18,18,2]

#plot bars with left x-coordinates [0,1,2,3,4], heights [num_champs]
plt.bar(range(len(schools)), num_champs)

plt.title("No. of men's basketball championships at the UAAP")
plt.ylabel("# of Championships")

#label x-axis with school names at bar centers
plt.xticks(range(len(schools)), schools)

plt.show()
```

## Scatterplots
A scatterplot is the right choice for visualizing the relationship between two paired sets of data.

```python
friends = [70, 65, 72, 63, 71, 64, 60, 64, 67]
minutes = [175, 170, 205, 120, 220, 130, 105, 145, 190]
labels = ['a','b','c','d','e','f','g','h','i']

plt.scatter(friends, minutes)

# label each point
for label, friend_count, minute_count in zip(labels, friends, minutes):
    plt.annotate(label,
            xy = (friend_count, minute_count), #put the label with its point
            xytext = (5 , -5),
            textcoords = 'offset points')
plt.title("Daily Minutes vs Number of Friends")
plt.xlabel("# of friends")
plt.ylabel("daily minutes spent on the site")
```
